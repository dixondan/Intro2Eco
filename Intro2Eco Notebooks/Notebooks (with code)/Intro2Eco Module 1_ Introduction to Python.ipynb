```
# We can create our own functions
# Let's say we want to find all the integers in a list of variables

def find_integers(my_list):
    # we first create an empty list to append to
    new_list = []
    #then do a for loop through all the items
    for i in my_list:
      if  type(i) is int: #if it meets out criteria, append it
        new_list.append(i)
      else: #if it does not meet the criteria, just pass / do nothing
        pass
    return new_list

my_list = [3, 7, 1,"123", 0, 1000, 9, 44.0, "celery"]

find_integers(my_list)

```

```
## Working with Pandas: We should clean the data before we plot it

winter_months = [6, 7, 8] #creating the winter month list
df2 = df[df.Month.isin(winter_months)] #subsetting only winter months
df2['rain'] = df2['Monthly Precipitation Total (millimetres)'] #creating a new column named rain 
df2_mean = df2.groupby("Year")['rain'].mean().reset_index() # grouping by Year, and taking the mean of rain for months 6, 7, 8 for each year (in case any years are missing a month)

## Important!: df2_mean is now the new data frame we will be working with. df and df2 are still stored in memory though!
```

**Plot graphs and maps**

```
# Make our first time series plot

plt.plot(df2_mean.Year, df2_mean.rain)
plt.xlabel("Year")
plt.ylabel("Mean winter rainfall (mm)")
plt.show()
```

```
# Let's plot the trend...

plt.plot(df2_mean.Year, df2_mean.rain)
plt.xlabel("Year")
plt.ylabel("Mean winter rainfall (mm)")

x = df2_mean.Year
y = df2_mean.rain

plt.scatter(x, y)

z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")
plt.show()
```

```
# Is it a significant trend?

import statsmodels.api as sm
X = df2_mean['Year']
y = df2_mean['rain']
X = sm.add_constant(X)
linear_model = sm.OLS(y, X)
result = linear_model.fit()
result.summary()
```

**Writing data, creating a shapefile and plotting a map**




```
# Make a new plot with our trend line then save it
plt.plot(df2_mean.Year, df2_mean.rain)
plt.xlabel("Year")
plt.ylabel("Mean winter rainfall (mm)")

equation = "y = 713.16 - 0.28x"
pval = "p = 0.001"

x = df2_mean.Year
y = df2_mean.rain

plt.scatter(x, y)

z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")
plt.text(1900, 90, equation) # 1900 is the x value and 90 is the y value on the plot
plt.text(1900, 75, pval) # 1900 is the x value 75 is the second y value on the plot
plt.title("Station: Cape Naturaliste")
#plt.savefig("/content/drive/My Drive/Colab Notebooks/Data/rainfall2/result.png", dpi = 200)
plt.show()
```

```
# Finally time to make a map! we can add some spatial units to the rainfall station and see it on a map

# First, we're going to make a shapefile from scratch

# What does a shapefile need? 
# 1. An x,y point
# 2. A coordinate reference system 
# That's it!


# To do this, we're going to create a pandas dataframe from scratch with just one row for our station

dfnew = pd.DataFrame(['Cape Naturaliste'], columns = ['Name'])
dfnew['latitude'] = -33.54 # I got these from the .txt file that comes with the download
dfnew['longitude'] = 115.02
dfnew.head()


```

```
# Next step we need to install a python package that does not come preinstalled
# that will let us work with our data frame as a shapefile attribute table

!pip install geopandas # You don't need to run !pip install every time. Only the first time running the cell. If you close the notebook or restart the notebook, you'll have to install again
import geopandas as gpd
```

```
# Now we have to make a GeoDataFrame with GeoPandas (pandas for shapefiles) convert these into a Geometry point that Python recognizes as a spatial point 

df2_xy = gpd.GeoDataFrame(dfnew, geometry = gpd.points_from_xy(dfnew.longitude, dfnew.latitude))
df2_xy.crs = {'init': 'epsg:4326'}

df2_xy.plot() # Our first spatial data plot in Python!
plt.show()
# Question, is this a map?
```

```
# Let's bring in some other spatial informaiton and make our map better

# Find the shapefile in your google drive

swwa_file = "/content/drive/My Drive/Intro2Eco/Data/sw_ecoregions/sw_ecoregions.shp"
swwa = gpd.read_file(swwa_file) # Read the swwa ecoregion shapefile into memory

# Now we want to subset our shapefile and make it cleaner (keeping only the five ecoregions we've worked with from yesterday)
swwa = swwa[swwa.SUB_CODE_7.isin(["JAF01", "JAF02","SWA01","WAR01","SWA02"])]

```

```
# Before we plot the point and the boundary shapefile, we need to make sure they are in the same coordinate reference system

print(swwa.crs)
print(df2_xy.crs)

# They are different!!
# If you're curious, make the plot with them with different crss and see what it looks like

# A major benefit of geopandas is we can do on-the-fly projections

df2_xy = df2_xy.to_crs({'init': 'epsg:3857'}) # find all epsg codes at this website it will be your best friend! https://spatialreference.org

df2_xy.head() # let's see our data frame
```
